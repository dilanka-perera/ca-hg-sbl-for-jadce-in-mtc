```python
import numpy as np

def surrogate_update_once(gamma_prev, Omega_ii, K, eta):
    """
    One surrogate update step for gamma_i with clipping.
    """
    a = 2*Omega_ii*gamma_prev + 2*K + 1.7
    a = max(0.0, a)

    if a == 0:
        return eta
    else:
        return (-1 + np.sqrt(1 + 4*a*eta)) / (2*a)

def cubic_roots(Omega_ii, K, eta):
    """
    Solve the cubic equation:
        2*Omega_ii*gamma^3 + 2*K*gamma^2 + gamma - eta = 0
    Return real roots (positive ones preferred).
    """
    coeffs = [2*Omega_ii, 2*K, 1, -eta]  # cubic coefficients
    roots = np.roots(coeffs)
    real_roots = roots[np.isreal(roots)].real
    return real_roots[real_roots > 0]  # keep only positive real roots
```

```python
import matplotlib.pyplot as plt

# Example usage
Omega_ii = -1
K = -0.2
eta = 1.5
gamma_prev = 0.5

# Sweep gamma0 values
gamma0_vals = np.linspace(0.01, 3, 300)
gamma_new_vals = [surrogate_update_once(g, Omega_ii, K, eta) for g in gamma0_vals]

# Get cubic roots
roots = cubic_roots(Omega_ii, K, eta)

# === Plot ===
plt.figure(figsize=(7,6))
plt.plot(gamma0_vals, gamma_new_vals, label=r"Surrogate update $f(\gamma_0)$", color="b")
plt.plot(gamma0_vals, gamma0_vals, "k--", label=r"$y=\gamma$")

# Mark cubic roots
for r in roots:
    plt.axvline(r, color="g", linestyle=":", label=f"Cubic root at {r:.2f}")
    plt.axhline(r, color="g", linestyle=":")

plt.xlabel(r"Initial $\gamma_0$")
plt.ylabel(r"Next $\gamma$")
plt.title("One-step Surrogate Update vs Cubic Roots")
plt.legend()
plt.grid(True)
plt.show()

print("Positive real cubic roots:", roots)

diff_vals = np.abs(np.array(gamma_new_vals) - gamma0_vals)

# === Plot ===
plt.figure(figsize=(7,6))
plt.plot(gamma0_vals, diff_vals, color="b", label=r"$|f(\gamma) - \gamma|$")
for r in roots:
    plt.axvline(r, color="g", linestyle=":", label=f"Cubic root at {r:.2f}")
plt.xlabel(r"Initial $\gamma$")
plt.ylabel(r"$|f(\gamma) - \gamma|$")
plt.title("Surrogate Update Error vs True Cubic Roots")
plt.legend()
plt.grid(True)
plt.show()

print("Positive real cubic roots:", roots)
```
