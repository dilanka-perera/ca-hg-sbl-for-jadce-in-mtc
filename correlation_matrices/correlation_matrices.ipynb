```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load saved data
from google.colab import drive
drive.mount('/content/drive')
data = np.load(f'{base_path}/general_model/data/general_model.npz', allow_pickle=True)
```

```python
# Access the saved data from the .npz file
all_DeviceLocations = data['all_DeviceLocations']
device_count = data['device_count']
num_samples = data['num_samples']
```

```python
first_20_devices = all_DeviceLocations[:, :400, :]
```

```python
loc = first_20_devices[401]
```

```python
distance_matrix = np.linalg.norm(loc[:, np.newaxis, :] - loc[np.newaxis, :, :], axis=2)
```

```python
kappa = 7
D = 40
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute correlation matrix
correlation_matrix = np.maximum((np.exp(-distance_matrix / kappa) - np.exp(-D / kappa)) / (1 - np.exp(-D / kappa)), 0)

# Compute inverse correlation matrix using pseudo-inverse
inverse_correlation_matrix = np.linalg.pinv(correlation_matrix)

# Define tick positions (0-based indexing but labeled as 1-based)
tick_positions = np.arange(4, 20, 5)  # Positions at 4, 9, 14, 19
tick_labels = tick_positions + 1  # Labels start from 1 instead of 0

plt.rcParams['font.family'] = 'serif'
plt.rcParams['text.usetex'] = False

# **Plot Correlation Matrix**
plt.figure(figsize=(6, 5))
cax1 = plt.imshow(correlation_matrix[:20, :20], cmap='RdBu', interpolation='nearest', vmin=-2, vmax=2)
cbar1 = plt.colorbar(cax1)
cbar1.ax.tick_params(labelsize=24)  # Adjust font size of colorbar ticks
plt.xticks(tick_positions, tick_labels, fontsize=24)
plt.yticks(tick_positions, tick_labels, fontsize=24)
plt.xlabel('Device Index', fontsize=24)
plt.ylabel('Device Index', fontsize=24)

# Save the correlation matrix figure with high DPI
plt.tight_layout()
plt.savefig(f'{base_path}/correlation_matrices/figures/correlation_matrix.eps', format='eps', bbox_inches='tight', dpi=300)

# **Plot Inverse Correlation Matrix**
plt.figure(figsize=(6, 5))
cax2 = plt.imshow(inverse_correlation_matrix[:20, :20], cmap='RdBu', interpolation='nearest', vmin=-2, vmax=2)
cbar2 = plt.colorbar(cax2)
cbar2.ax.tick_params(labelsize=24)  # Adjust font size of colorbar ticks
plt.xticks(tick_positions, tick_labels, fontsize=24)
plt.yticks(tick_positions, tick_labels, fontsize=24)
plt.xlabel('Device Index', fontsize=24)
plt.ylabel('Device Index', fontsize=24)

# Save the inverse correlation matrix figure with high DPI
plt.tight_layout()
plt.savefig(f'{base_path}/correlation_matrices/figures/inverse_covariance_matrix.eps', format='eps', bbox_inches='tight', dpi=300)

# Show the plots
plt.show()
plt.close()
```

```python
# === Build Precision Matrix ===
precision_matrix = -np.maximum(
    (np.exp(-distance_matrix / kappa) - np.exp(-D / kappa)) / (1 - np.exp(-D / kappa)),
    0
)
np.fill_diagonal(precision_matrix, 10)
precision_matrix = precision_matrix + 0.1

# === Covariance Matrix (pseudo-inverse of precision) ===
covariance_matrix = np.linalg.pinv(precision_matrix)

# === Tick positions and labels ===
tick_positions = np.arange(4, 20, 5)  # Positions at 4, 9, 14, 19
tick_labels = tick_positions + 1      # Labels start from 1 instead of 0

plt.rcParams['font.family'] = 'serif'
plt.rcParams['text.usetex'] = False

# === Plot Distance Matrix ===
plt.figure(figsize=(6, 5))
cax0 = plt.imshow(distance_matrix[:20, :20], cmap='viridis', interpolation='nearest')
cbar0 = plt.colorbar(cax0)
cbar0.ax.tick_params(labelsize=24)
plt.xticks(tick_positions, tick_labels, fontsize=24)
plt.yticks(tick_positions, tick_labels, fontsize=24)
plt.xlabel('Device Index', fontsize=24)
plt.ylabel('Device Index', fontsize=24)
plt.tight_layout()

# === Plot Precision Matrix ===
plt.figure(figsize=(6, 5))
cax1 = plt.imshow(precision_matrix[:20, :20], cmap='RdBu', interpolation='nearest', vmax = np.max(np.abs(precision_matrix)), vmin = -np.max(np.abs(precision_matrix)))
cbar1 = plt.colorbar(cax1)
cbar1.ax.tick_params(labelsize=24)
plt.xticks(tick_positions, tick_labels, fontsize=24)
plt.yticks(tick_positions, tick_labels, fontsize=24)
plt.xlabel('Device Index', fontsize=24)
plt.ylabel('Device Index', fontsize=24)
plt.tight_layout()

# === Plot Covariance Matrix ===
plt.figure(figsize=(6, 5))
cax2 = plt.imshow(covariance_matrix[:20, :20], cmap='RdBu', interpolation='nearest', vmax = np.max(np.abs(covariance_matrix)), vmin = -np.max(np.abs(covariance_matrix)))
cbar2 = plt.colorbar(cax2)
cbar2.ax.tick_params(labelsize=24)
plt.xticks(tick_positions, tick_labels, fontsize=24)
plt.yticks(tick_positions, tick_labels, fontsize=24)
plt.xlabel('Device Index', fontsize=24)
plt.ylabel('Device Index', fontsize=24)
plt.tight_layout()

# === Show plots ===
plt.show()
plt.close()
```
